```python
import matplotlib.pyplot as plt 
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as shc
import pandas as pd 
```

```python
df = pd.read_csv(r"C:\DataScience\Machine Learning Algorithms\Hierarchical Clustering\hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv")
display(df.head())
print(df.shape)
```

```python
df = df.iloc[:,3:5].values
```

```python
plt.figure(figsize=(10,7))
plt.title('Customer Dendograms')
dend = shc.dendrogram(shc.linkage(df, method='ward'))
```

```python
cluster = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
label_ = cluster.fit_predict(df)
```

```python
label_
```

```python
plt.scatter(df[:,0], df[:,1], c=cluster.labels_, cmap='summer')
```
