```python
import numpy as np
from sklearn.datasets import make_circles
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import cross_val_score
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X,y = make_circles(n_samples=500, factor=0.1, noise=0.32, random_state=42)
```

```python

sns.scatterplot(x=X[:,0], y=X[:,1], hue=y)
```

```python
# create a objet -> fit and check the cross_val_score without any hyper-parameter tune

abc = AdaBoostClassifier()

print(np.mean(cross_val_score(abc, X,y, scoring='accuracy', cv=10)))
```

```python
abc.fit(X,y)
```

```python
def plot_decision_boundary(clf):
    plt.figure(figsize=(9, 6))
    x_range = np.linspace(X.min(), X.max(), 100)
    xx1, xx2 = np.meshgrid(x_range, x_range)
    y_hat = clf.predict(np.c_[xx1.ravel(), xx2.ravel()])
    y_hat = y_hat.reshape(xx1.shape)
    plt.contourf(xx1, xx2, y_hat, alpha=0.2)
    plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis', alpha =.7)
    plt.title("Adaboost Classifier")
    plt.show()
```

```python
plot_decision_boundary(abc)
```

```python
# after tuning hyper-parameter

abc = AdaBoostClassifier(n_estimators=700, learning_rate=0.1)
abc.fit(X,y)
print(np.mean(cross_val_score(abc, X,y, scoring='accuracy', cv=10)))

plot_decision_boundary(abc)
```

# find best parameter by using "GridSearchCv"

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = dict()
grid['n_estimators']=[50,200,500,800,1200,1500,2000]
grid['learning_rate'] = [0.01,0.07,0.1,0.3,0.7,1.0]
grid['algorithm'] = ['SAMME','SAMME.R']

grid_search = GridSearchCV(estimator=AdaBoostClassifier(), param_grid= grid, cv=10, n_jobs=-1, scoring='accuracy')

# execute the grid Search
grid_result = grid_search.fit(X,y)

# summarize the best score and configuration
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```
