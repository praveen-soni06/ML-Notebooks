# Drawing Shape and text on image

```python
## creation our own image

img = np.zeros((512,512,3))
cv2.imshow('own image', img)
cv2.waitKey(0)
```

```python
## ractangle create

cv2.rectangle(img, pt1 = (100,150), pt2 = (300,500), color = (0,255,0), thickness = 1)
cv2.imshow('own image', img)
cv2.waitKey(0)
```

```python
## circle create

cv2.circle(img, center = (190,450),radius = 50, color = (18,255,25), thickness = -1)
cv2.imshow('own image', img)
cv2.waitKey(0)
```

```python
## line create

cv2.line(img, pt1 = (150,238), pt2 = (340,560), color = (0,255,0), thickness = 5)
cv2.imshow('own image', img)
cv2.waitKey(0)
```

```python
##  text(font)  add into the image

cv2.putText(img, org=(10,300), fontScale=2, color=(0,255,255), thickness = 2, lineType = cv2.LINE_AA,text="praveen openCV",fontFace =cv2.FONT_ITALIC)
cv2.imshow("Puting text", img)
cv2.waitKey(0)
```

# working with opencv events

```python
img = cv2.imread(r"C:\DataScience\Matplotlib_YT\sample1.png")

def draw(event,x,y,flags,params):
    if event == 1:
        cv2.circle(img, center=(x,y), radius = 10,color = (19,34,66), thickness = -1)
        coord_text = f"({x},{y})"
        
        # Put coordinates above the circle (slightly shifted upward)
        cv2.putText(img, coord_text,(x-37,y-19),  # shift left and up
                    fontFace=cv2.FONT_HERSHEY_SIMPLEX, 
                    fontScale=0.5, 
                    color=(0, 0, 255),  # red text
                    thickness=1)
        
cv2.namedWindow(winname='own image')
cv2.setMouseCallback('own image', draw)

while True:
    cv2.imshow('own image', img)
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break

cv2.destroyAllWindows()
```

```python
##  making ractangle (click & drag)

img = cv2.imread(r"C:\DataScience\Matplotlib_YT\sample1.png")

flag = False
ix = -1
iy = -1

def draw(event,x,y,flags,params):

    global flag, ix, iy
    if event == cv2.EVENT_LBUTTONDOWN:

        flag = True
        ix = x
        iy = y
        
    elif event == cv2.EVENT_MOUSEMOVE:
        if flag == True:

            cv2.rectangle(img, pt1=(ix,iy), pt2=(x,y), color = (123,156,231), thickness = -1 )

    elif event == cv2.EVENT_LBUTTONUP:
        
        flag = False
        cv2.rectangle(img, pt1=(ix,iy), pt2=(x,y), color = (123,156,231), thickness = -1 )

        
        coord_text = f"({ix},{iy})"
        
        cv2.putText(img, coord_text,(ix-37,iy-19),  # shift left and up
                    fontFace=cv2.FONT_HERSHEY_SIMPLEX, 
                    fontScale=0.5, 
                    color=(0, 0, 255),  # red text
                    thickness=1)
        
cv2.namedWindow(winname='own image')
cv2.setMouseCallback('own image', draw)

while True:
    cv2.imshow('own image', img)
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break

cv2.destroyAllWindows()
```

# Building a croping tool

```python
## croping tool

img = cv2.imread(r"C:\Users\prabh\Downloads\wallpaperflare.com_wallpaper.jpg")
img = cv2.resize(img, (700,450))
display_img = img.copy()

flag = False
ix = -1
iy = -1
i = 1

def crop(event, x,y, flags,params):

    global flag,ix,iy,i, display_img
    
    if event == cv2.EVENT_LBUTTONDOWN:

        flag = True
        ix = x
        iy = y

    elif event == cv2.EVENT_MOUSEMOVE:

        if flag == True:
            display_img = img.copy()
            cv2.rectangle(display_img, pt1=(ix,iy), pt2=(x,y), color = (0,0,255), thickness = 2)
            
    elif event == cv2.EVENT_LBUTTONUP:
        flag = False
        
        roi = img[iy:y,ix:x]  ## use for croping the image

        if roi.size > 0:
            
            cv2.imwrite(f'croped_Simage{i}.png', roi)
            print(f'croped_Simage{i}.png')
            i += 1


        display_img = img.copy()   ## reset the image copy


cv2.namedWindow(winname = 'naruto image')
cv2.setMouseCallback('naruto image', crop)


while True:

    cv2.imshow('naruto image', display_img)

    if cv2.waitKey(1) & 0xFF == ord('x'):
        break

cv2.destroyAllWindows()
```

#  Working with videos using openCV


```python
##   recording the video using laptop camera 

cap = cv2.VideoCapture(0)   # laptop camera

fourcc = cv2.VideoWriter_fourcc(*'XVID')   # codec
out = cv2.VideoWriter('output.avi', fourcc, 20.0, (640,480), isColor=False)  # save in grayscale

while True:
    ret, frame = cap.read()   # fixed spelling
    
    if not ret:
        break

    img_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # convert to gray
    out.write(img_gray)   # write frame to file

    cv2.imshow("webcam", img_gray)  # display

    if cv2.waitKey(1) & 0xFF == ord('x'):  # press x to quit
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```
