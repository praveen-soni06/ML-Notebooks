```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans
import pandas as pd
```

```python
df = pd.read_csv(r"C:\DataScience\Machine Learning Algorithms\KMeans\student_clustering.csv")
```

# find n_cluster by using WCSS

```python
wcss = []

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit(df)
    wcss.append(km.inertia_)
```

```python
# ploting the wcss graph

plt.plot(range(1,11), wcss)
```

```python
# final code of kmeans

X = df.iloc[:,:].values
km = KMeans(n_clusters=4)
y_means = km.fit_predict(X)
```

```python
# ploting the clusters

plt.scatter(x=X[y_means == 0,0], y=X[y_means == 0,1], c='red')
plt.scatter(x=X[y_means == 1,0], y=X[y_means == 1,1], c='blue')
plt.scatter(x=X[y_means == 2,0], y=X[y_means == 2,1], c='pink')
plt.scatter(x=X[y_means == 3,0], y=X[y_means == 3,1], c='yellow')
```
