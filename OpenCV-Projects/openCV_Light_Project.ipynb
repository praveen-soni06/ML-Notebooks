```python
## eye detection inside the face rectengle

face_classifier = cv2.CascadeClassifier(r"C:\DataScience\Computer-Vision-Tutorial-master\Computer-Vision-Tutorial-master\Haarcascades\haarcascade_frontalface_default.xml")
eye_classifier = cv2.CascadeClassifier(r"C:\DataScience\Computer-Vision-Tutorial-master\Computer-Vision-Tutorial-master\Haarcascades\haarcascade_eye.xml")

img = cv2.imread(r"M:\IMG_20240127_160804.jpg")
h,w = img.shape[:2]
new_shape = cv2.resize(img, (700, int(h * 700 / w)))

gray = cv2.cvtColor(new_shape, cv2.COLOR_BGR2GRAY)

faces = face_classifier.detectMultiScale(gray, 1.1,3)

if len(faces) == 0:
    print("no face is found")

else:
    for (x,y,w,h) in faces:
        cv2.rectangle(new_shape, pt1=(x,y), pt2 = (x+w, y+h), color= (127,0,255), thickness = 2)
        
        # extact ROI for eyes
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = new_shape[y:y+h, x:x+w]
        
        # detect eye inside the face ROI
        eyes = eye_classifier.detectMultiScale(roi_gray)

        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(roi_color, (ex,ey), (ex+ew,ey+eh), (255,255,0), 2)

    # show result once after all detections
    cv2.imshow('eye detection', new_shape)
    cv2.waitKey(0)
    

cv2.destroyAllWindows()
```

# face and eye Detction from videos

```python
##  face recognition

face_cascade = cv2.CascadeClassifier(r"C:\DataScience\Computer-Vision-Tutorial-master\Computer-Vision-Tutorial-master\Haarcascades\haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier(r"C:\DataScience\Computer-Vision-Tutorial-master\Computer-Vision-Tutorial-master\Haarcascades\haarcascade_eye.xml")

def detect(gray, frame):
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:

        cv2.rectangle(frame, (x,y), (x+w, y+h), (255, 0,0), 2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_frame = frame[y:y+h, x:x+w]

        eyes = eye_cascade.detectMultiScale(roi_gray, 1.1,3)
        for (ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_frame, (ex,ey), (ex+ew, ey+eh), (0,255,0), 2)
    return frame

video_capture = cv2.VideoCapture(0)

while True:
    _ , frame = video_capture.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    canvas = detect(gray, frame)
    cv2.imshow("video", canvas)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video_capture.release()
cv2.destroyAllWindows()
```

##  Pedestrian (person full body) Detection using OpenCV



```python
# Load Haarcascade classifier
body_classifier = cv2.CascadeClassifier(r"C:\DataScience\Computer-Vision-Tutorial-master\Computer-Vision-Tutorial-master\Haarcascades\haarcascade_fullbody.xml")

cap = cv2.VideoCapture(r"C:\Users\prabh\Downloads\stock-footage-street-camera-view-of-people-walking-in-pedestrian-zone-high-angle-view.webm")

while cap.isOpened():

    ret, frame = cap.read()

    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect bodies
    bodies = body_classifier.detectMultiScale(gray, 1.1,3)

    # Draw rectangles
    for (x,y,w,h) in bodies:

        cv2.rectangle(frame, (x,y),(x+w, y+h), (24,255,98), 2)
        cv2.imshow('body dectection', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

# car detection using openCV

```python


car_classifier = cv2.CascadeClassifier(r"C:\DataScience\Computer-Vision-Tutorial-master\Computer-Vision-Tutorial-master\Haarcascades\haarcascade_car.xml")

cap = cv2.VideoCapture(r"C:\Users\prabh\Downloads\27260-362770008_large.mp4")

while cap.isOpened():

    ret, frame = cap.read()

    if not ret:
        break
        
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    cars = car_classifier.detectMultiScale(gray, 1.1, 3)

    for (x,y,w,h) in cars:

        cv2.rectangle(frame, (x,y), (x+w, y+h), (104,255,198),2)
        cv2.imshow("car detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()

cv2.destroyAllWindows()

    
```
