```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
```

```python
x,y = make_classification(n_samples=200, n_features=2, n_informative=1, n_redundant=0,
                         n_classes=2, n_clusters_per_class=1, random_state=41, hypercube=False, class_sep = 10)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0], x[:,1], c=y, cmap='winter', s=100)
plt.show()
```

```python
def perceptron(x,y):
    x= np.insert(x,0,1,axis=1)    # inset a new column at '0' index in x which value is '1'
    weights = np.ones(x.shape[1])  # create a new array
    lr = 0.1
    epochs = 1000

    for i in range(epochs):
        # pick random point 
        j = np.random.randint(0,200)
        # calculating y_hat
        y_hat = step(np.dot(x[j], weights))
        # changing weights
        weights = weights + (lr*(y[j]-y_hat)*x[j])

    return weights[0], weights[1:]
```

```python
def step(z):
    # return 1 if z>0 else 0  # use step function
    return 1/(1+ np.exp(-z))  # use sigmoid function
```

```python
intercept_, coef_ = perceptron(x,y)
```

```python
intercept_
```

```python
coef_
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input +b
```

```python
#  by using of step function

plt.figure(figsize=(10,6))
plt.plot(x_input, y_input,color='red', linewidth=3)
plt.scatter(x[:,0], x[:,1],c=y,cmap='winter', s=100)
plt.ylim(-3,2)
```
