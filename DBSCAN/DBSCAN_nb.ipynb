```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px 
from sklearn.cluster import DBSCAN
import pandas as pd
```

```python
df = pd.read_csv(r"C:\DataScience\Machine Learning Algorithms\DBSCAN\dbscan_10_clusters_dataset.csv")
print(df.shape)
display(df.head())
```

```python
dbscan = DBSCAN(eps=0.7, min_samples=5, n_jobs=-1)
```

```python
X = df.loc[:,['Feature_1','Feature_2']]
```

```python
dbscan.fit_predict(X)
```

```python
labels = dbscan.labels_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=df['Feature_1'], y=df['Feature_2'], c=labels)
plt.title('DBSCAN Cluster Graph')
plt.xlabel('Feature_1')
plt.ylabel('Feature_2')
```
